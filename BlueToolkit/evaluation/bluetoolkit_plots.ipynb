```python
## From Specification to a car
import statistics

differences = []
for year, vers in zip(years, versions):
    #print(str(year) + " " + str(vers) + " " + str(year - versions_released[vers]))
    differences.append(year - versions_released[vers])

differences.sort()
diff_mean = statistics.mean(differences)
diff_median = statistics.median(differences)
diff_stdv = statistics.stdev(differences)
diff_mode = statistics.multimode(differences)
diff_variance = statistics.variance(differences)

print(str(diff_mean) + " " + str(diff_median) + " " + str(diff_stdv) + " " + str(diff_mode) + " " + str(diff_variance))
print(differences)
```

```python
from collections import Counter
datax = zip(years, versions)
datax = sorted(datax, key=lambda x: x[0])
counter = Counter(datax)

clusters = {}
MAX_YEAR = 2024
MIN_YEAR = 2015

for year in range(MIN_YEAR, MAX_YEAR):
    clusters[year] = {}
    for version in version_years.keys():
        clusters[year][version] = 0

for (year, version), count in counter.items():
    clusters[year][version] += count



x_indexes = [i for i in range(0, (MAX_YEAR - MIN_YEAR) * 11 + 1, 1)]
x_values = []
colors = []
x_tickvals = []
x_tickvals_text = []

for year, versionss in clusters.items():
    x_values.append(0)
    colors.append('blue')
    x_tickvals.append(year)
    for version, count in versionss.items():
        if count != 0:
            x_values.append(float(count))
            colors.append(versions_colors[version])
            x_tickvals.append(version)

x_values.append(0)
colors.append('blue')
x_tickvals.append(MAX_YEAR)

for tick in x_tickvals:
    if tick >= MIN_YEAR and tick <= MAX_YEAR:
        x_tickvals_text.append(f"<b>{tick}</b>")
    else:
        x_tickvals_text.append(tick)


fig = go.Figure()

# Add bars for each year nearby

fig.add_trace(go.Bar(
    x=x_indexes,
    y=x_values,
    textposition='auto',
    marker_color=colors,
    name=f'Year {year}'
))

# Update layout
fig.update_layout(
    title='BT version occurances across years',
    xaxis=dict(title='Year / BT version', tickmode='array', tickvals=[i for i in range(0, len(x_tickvals))], ticktext=x_tickvals_text),
    yaxis=dict(title='Occurrences'),
    barmode='stack'
)


fig.show()

```

```python
group_clusters = {}

for year in range(MIN_YEAR, MAX_YEAR, 5):
    group_clusters[year] = {}
    for version in version_years.keys():
        group_clusters[year][version] = 0

print(group_clusters)

for (year, version), count in counter.items():
    if year <= 2019:
        group_clusters[MIN_YEAR][version] += count
    else:
        group_clusters[MIN_YEAR+5][version] += count


x_indexes = [i for i in range(0, (MAX_YEAR - MIN_YEAR) * 11 + 1, 1)]
x_values = []
colors = []
x_tickvals = []
x_tickvals_text = []

for year, versionss in group_clusters.items():
    x_values.append(0)
    colors.append('blue')
    x_tickvals.append(year)
    for version, count in versionss.items():
        if count != 0:
            x_values.append(float(count))
            colors.append(versions_colors[version])
            x_tickvals.append(version)

x_values.append(0)
colors.append('blue')
x_tickvals.append(MAX_YEAR)

for tick in x_tickvals:
    if tick >= MIN_YEAR and tick <= MAX_YEAR:
        x_tickvals_text.append(f"<b>{tick}</b>")
    else:
        x_tickvals_text.append(tick)


print(x_indexes)
print(x_values)
print(colors)

fig = go.Figure()

# Add bars for each year nearby

fig.add_trace(go.Bar(
    x=x_indexes,
    y=x_values,
    textposition='auto',
    marker_color=colors,
    name=f'Year {year}'
))

# Update layout
fig.update_layout(
    title='BT version occurances across 2015-2019 and 2020-2023',
    xaxis=dict(title='Year / BT version', tickmode='array', tickvals=[i for i in range(0, len(x_tickvals))], ticktext=x_tickvals_text),
    yaxis=dict(title='Occurrences'),
    barmode='stack'
)


fig.show()
```

```python



DESIGN_EXPLOITS = ['custom_method_confusion_check', 'custom_nino_check', 'internalblue_CVE_2018_5383_Invalid_second',  'braktooth_knob', 'steal_contacts', 'custom_legacy_pairing_second_check']

RECON_EXPLOITS = ['fast_reboot', 'only_vehicle_can_connect', 'reconnaissance_SC_supported', 'always_pairable','reconnaissance_SSP_supported','reconnaissance_possible_BLUR']

IMPLEMENTATION_EXPLOITS = ['bleedingtooth_badchoice_cve_2020_12352', 'bleedingtooth_badkarma_cve_2020_12351',
                     'bleedingtooth_badvibes_cve_2020_24490',
                    'blueborne_CVE_2017_1000250', 'blueborne_CVE_2017_0785', 'blueborne_CVE_2017_1000251','my_name_is_keyboard','custom_numeric_wrong_implementation','invalid_feature_page_execution','au_rand_flooding', 'feature_response_flooding','feature_req_ping_pong', 'paging_scan_disable', 'wrong_encapsulated_payload',
               'duplicated_iocap', 'lmp_overflow_2dh1','lmp_overflow_dm1','lmp_auto_rate_overflow','sdp_oversized_element_size','lmp_invalid_transport',
               'lmp_max_slot_overflow','repeated_host_connection','duplicated_encapsulated_payload','sdp_unkown_element_type',
               'truncated_sco_link_request','truncated_lmp_accepted','invalid_setup_complete', 'invalid_max_slot', 'invalid_timing_accuracy','check_nino_vehicle']

### possible_BLUR - basically: if device supports BLE (that's it).
FILTER_OUT = ['internalblue_CVE_2018_5383_Invalid', 'internalblue_knob']

maybe = 'reconnaissance_SSP_supported'

CVE_2018_19860_EXPLOITS = ['internalblue_CVE_2018_19860_16_0b', 'internalblue_CVE_2018_19860_20_17', 'internalblue_CVE_2018_19860_0a_00']


opel_counted = []
tesl_counted = []


data_extraction = []
for report in data:
    amount_of_tested_exploits = 0
    amount_of_vuln_design_exploits = 0
    amount_of_vuln_recon_exploits = 0
    amount_of_vuln_exploits = 0
    amount_of_vuln_implementation_exploits = 0


    report_exploits = report['done_exploits']
    report_exploits.extend(report['manual_exploits'])
    print(report_exploits)
    cve2018_19860_list_vulnerable = False
    for exploit in report_exploits:
        if exploit['name'] not in FILTER_OUT:
            if exploit['name'] == maybe:
                amount_of_tested_exploits += 2
                if exploit['data'] == "SSP not supported":
                    amount_of_vuln_exploits += 2
                    amount_of_vuln_recon_exploits += 2
                elif exploit['data'] == "SSP supported, weak cryptography is used due to BT version <":
                    amount_of_vuln_exploits += 1
                    amount_of_vuln_recon_exploits += 1
                elif exploit['data'] == "SSP supported, secure cryptography is used, there might be a problem with Message Intergrity":
                    pass
                else:
                    print("error")
                    print(exploit)
            elif exploit['name'] in CVE_2018_19860_EXPLOITS:
                if exploit["code"] == 2:
                    cve2018_19860_list_vulnerable = True
            else:
                amount_of_tested_exploits += 1
                if exploit["code"] == 2:
                    amount_of_vuln_exploits += 1
                    if exploit['name'] in DESIGN_EXPLOITS:
                        amount_of_vuln_design_exploits += 1
                    elif exploit['name'] in RECON_EXPLOITS:
                        amount_of_vuln_recon_exploits += 1
                    elif exploit['name'] in IMPLEMENTATION_EXPLOITS:
                        amount_of_vuln_implementation_exploits += 1

                    else:
                        print(exploit['name'])
    amount_of_tested_exploits += 1
    if cve2018_19860_list_vulnerable:
        amount_of_vuln_exploits += 1
        amount_of_vuln_implementation_exploits += 1

    data_extraction.append([report["bt_version"], report['manufacturer'], report['vehicle_name'], report['vehicle manufacturer'], report['parent_company'], report['year'], amount_of_tested_exploits, amount_of_vuln_exploits, amount_of_vuln_recon_exploits, amount_of_vuln_design_exploits, amount_of_vuln_implementation_exploits])

```

```python
import matplotlib.pyplot as plt

# Reload the new CSV file
new_file_path = '/content/sample_data/study_car.csv'
df = pd.read_csv(new_file_path)
# Count the number of cars for each Bluetooth manufacturer
manufacturer_counts = df['BT Manufacturer'].str.split().str[0].value_counts()
# Aggregate vulnerability data by Bluetooth manufacturer
bt_manufacturer_vulnerabilities = df.groupby(df['BT Manufacturer'].str.split().str[0])[
    ['Vulnerabilities', 'Recon Vulnerabilities', 'Design Vulnerabilities', 'Implementation Vulnerabilities']
].mean()
# Rename 'Implementation Vulnerabilities' to 'Impl. Vulnerabilities'
bt_manufacturer_vulnerabilities = bt_manufacturer_vulnerabilities.rename(
    columns={'Implementation Vulnerabilities': 'Impl. Vulnerabilities'}
)
# Rename indices in the vulnerabilities dataframe with car counts and update 'Texas' to 'Texas Inc.'
bt_manufacturer_vulnerabilities.index = [f"{manufacturer.replace('Texas', 'Texas Inc.')} ({manufacturer_counts[manufacturer]})"
    for manufacturer in bt_manufacturer_vulnerabilities.index]
# Plot vulnerabilities by Bluetooth manufacturer with updated labels
plt.figure(figsize=(12, 6))
# Plot each category of vulnerabilities
bt_manufacturer_vulnerabilities.plot(kind='bar', stacked=False, figsize=(12, 6))
# Labels and title with increased font sizes
plt.xlabel('Bluetooth Manufacturer', fontsize=16)
plt.ylabel('Average # of Vulnerabilities', fontsize=16)
plt.title('Vulnerabilities by Bluetooth Manufacturer', fontsize=18)
plt.xticks(rotation=45, ha='right', fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
# Show plot
plt.show()



# Reload the new CSV file
new_file_path = '/content/sample_data/study_car.csv'
df = pd.read_csv(new_file_path)
# Exclude the 'Sum' row if present in the dataset
df = df[df['BT Version'] != 'Sum']
# Ensure that versions '3', '4", and '5' are displayed as '3.0", '4.0', and '5.0"
df['BT Version'] = df['BT Version'].replace({'3': '3.0', '4': '4.0', '5': '5.0'})
# Extract Bluetooth version and compute average vulnerabilities again
bt_vulnerabilities = df.groupby('BT Version')[
    ['Vulnerabilities', 'Recon Vulnerabilities', 'Design Vulnerabilities', 'Implementation Vulnerabilities']
].mean()
# Sort versions numerically
bt_vulnerabilities = bt_vulnerabilities.sort_index()
# Map Bluetooth versions to equidistant x-axis positions
used_bt_versions = bt_vulnerabilities.index.astype(str).tolist()
bt_version_mapping = {version: i for i, version in enumerate(used_bt_versions)}
x_positions = list(bt_version_mapping.values())
# Define color groups for major Bluetooth version groupings with distinct shading for 2.1 and 3.0
major_groups_colors = {
    'BT 2.1': ('deepskyblue', ['2.1']),
    'BT 3.0': ('royalblue', ['3.0']),
    'BT 4.x': ('lightgreen', ['4.0', '4.1', '4.2']),
    'BT 5.x': ('lightcoral', ['5.0', '5.1', '5.2']),
}
# Plot vulnerabilities by Bluetooth version after excluding 'Sum'
plt.figure(figsize=(10, 5))
# Apply shading with distinct colors for 2.1 and 3.0, plus the other groups
for group, (color, versions) in major_groups_colors.items():
    valid_versions = [v for v in versions if v in bt_version_mapping]
    if valid_versions:
        x_min = bt_version_mapping[valid_versions[0]] - 0.5
        x_max = bt_version_mapping[valid_versions[-1]] + 0.5
        plt.axvspan(x_min, x_max, color=color, alpha=0.3)
# Plot data using the equidistant positions
plt.plot(x_positions, bt_vulnerabilities['Vulnerabilities'], marker='o', linestyle='-', label='Total Vulnerabilities')
plt.plot(x_positions, bt_vulnerabilities['Recon Vulnerabilities'], marker='s', linestyle='--', label='Reconnaissance')
plt.plot(x_positions, bt_vulnerabilities['Design Vulnerabilities'], marker='^', linestyle='-.', label='Design')
plt.plot(x_positions, bt_vulnerabilities['Implementation Vulnerabilities'], marker='d', linestyle=':', label='Implementation')
# Adjust x-axis labels and increase font size
plt.xticks(x_positions, used_bt_versions, fontsize=12)
plt.xlabel('Bluetooth Version', fontsize=14)
plt.ylabel('Average # of Vulnerabilities', fontsize=14)
plt.title('Vulnerabilities by Bluetooth Version', fontsize=16)
# Extend y-axis to 14 for better visualization
plt.ylim(0, 14)
# Place legend above the plotted lines with increased font size
plt.legend(loc='upper right', fontsize=12)
plt.grid(True)
plt.show()


```
